Задча № 3, Вариант № 7:
\begin{equation*}
 \begin{cases}
   \frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2} = -2u,
   \\
   u(0,y)=\cos{y},
   \\
   u(\frac{\pi}{2},y) = 0,
   \\
   u(x,0)=\cos{x}
   \\
   u(x,\frac{\pi}{2})=0.
 \end{cases}
\end{equation*}

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

Аналитическое решение:
\begin{equation*}
    U(x,y)=\cos x\cos y
\end{equation*}

```python
from matplotlib import cm
L = math.pi / 2
n = 100
dl = (L)/(n-1)
x = np.arange(0,L+L/(n+1),dl)
y = np.arange(0,L+L/(n+1),dl)
x,y = np.meshgrid(x,y)
u = np.cos(x)*np.cos(y)
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(x, y, u, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.view_init(30,0)
plt.draw()
```

Конечно разностная апроксимация:
\begin{equation*}
 \begin{cases}
   \frac{u_{i-1,j}-2u_{i,j}+u_{i+1,j}}{h_x^2}+\frac{u_{i,j-1}-2u_{i,j}+u_{i,j+1}}{h_y^2} + O(h_x^2,h_y^2) = -2u_{i,j},
   \\
   u_{0,j}=\cos{j*h_y},
   \\
   u_{N_x,j} = 0,
   \\
   u_{i,0}=\cos{i*h_x}
   \\
   u_{i,N_y} = 0,
 \end{cases}
\end{equation*}
